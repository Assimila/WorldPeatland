```python
red_band = arr_b04[idx]
green_band = arr_b03[idx]
blue_band = arr_b02[idx]
nir_band = arr_b08[idx]

# def scale_min_max(array):
#     return (array - np.nanmin(array)) / (np.nanmax(array) - np.nanmin(array))

# red_normalized = scale_min_max(red_band)
# green_normalized = scale_min_max(green_band)
# blue_normalized = scale_min_max(blue_band)
# nir_normalized = scale_min_max(nir_band)

def scale_and_stretch(data, dst_min=0, dst_max=255, src_min=0, src_max=2700):

    # Scale and stretch the data
    scaled_data = np.clip((data - src_min) / (src_max - src_min) * (dst_max - dst_min) + dst_min, dst_min, dst_max)
    
    return (scaled_data.astype(np.uint8))

red_normalized = scale_and_stretch(red_band, dst_min=0, dst_max=255, src_min=0, src_max=3250)
green_normalized = scale_and_stretch(green_band, dst_min=0, dst_max=255, src_min=0, src_max=3000)
blue_normalized = scale_and_stretch(blue_band, dst_min=0, dst_max=255, src_min=0, src_max=3250)

rgb_normalized = np.dstack((red_normalized, green_normalized, blue_normalized))

nir_normalized = scale_and_stretch(nir_band, dst_min=0, dst_max=255, src_min=0, src_max=5500)
red_normalized = scale_and_stretch(red_band, dst_min=0, dst_max=255, src_min=0, src_max=5500)
green_normalized = scale_and_stretch(green_band, dst_min=0, dst_max=255, src_min=0, src_max=5500)

false_normalized = np.dstack((nir_normalized, red_normalized, green_normalized))
```

```python
def label_utm_grid(ax):
    ''' 
    Warning: should only use with small area UTM maps 
    source: https://github.com/SciTools/cartopy/issues/491
    '''    
    for val,label in zip(ax.get_xticks(), ax.get_xticklabels()):
        label.set_text(str(val))
        label.set_position((val,0))  

    for val,label in zip(ax.get_yticks(), ax.get_yticklabels()):   
        label.set_text(str(val))
        label.set_position((0,val))  

    plt.tick_params(bottom=True,top=True,left=True,right=True,
            labelbottom=True,labeltop=False,labelleft=True,labelright=False)

    ax.xaxis.set_visible(True)
    ax.yaxis.set_visible(True)
```

```python
# get lat lon from any xarray above to plot it correctly in its projection
lat = xr_b04['latitude']
lon = xr_b04['longitude']

fig, axes = plt.subplots(1, 2, figsize=(10, 4), subplot_kw={'projection': utm_crs})

# Plot RGB image
axes[0].imshow(rgb_normalized, extent=[lon.min(), lon.max(), lat.min(), lat.max()], transform=utm_crs)
axes[0].set_title('True Color Composite (RGB)')
label_utm_grid(axes[0])

# Plot False Color image
axes[1].imshow(false_normalized, extent=[lon.min(), lon.max(), lat.min(), lat.max()], transform=utm_crs)
axes[1].set_title('False Color Composite (NIR, R, G)')
label_utm_grid(axes[1])

fig.suptitle(f'{site}')
plt.tight_layout()
plt.savefig(f'/workspace/WorldPeatland/visualisation/TrueFalse/{site}_True_False_comp.png', dpi=200)
plt.show()
```
